```python
# Grouping dan menghitung total pendapatan berdasarkan tahun dan cabang olahraga
total_earnings_by_sport_by_year = df.groupby(['Year', 'Sport'])['Earnings ($ million)'].sum()

# Mengurutkan data berdasarkan total pendapatan dalam setiap tahun
total_earnings_by_sport_by_year_sorted = total_earnings_by_sport_by_year.reset_index().sort_values(['Year', 'Earnings ($ million)'], ascending=[True, False])

# Memilih 3 cabang olahraga dengan pendapatan terbanyak setiap tahunnya
top_3_sports_by_year = total_earnings_by_sport_by_year_sorted.groupby('Year').head(3)

# Menampilkan hasil
print(top_3_sports_by_year)
```

```python
df_total_earnings = total_earnings_by_sport_by_year.reset_index()

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_total_earnings, x='Year', y='Earnings ($ million)', hue='Sport')
plt.title("Scatter Plot: Total Earnings by Sport and Year", fontsize=15)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Earnings ($ million)', fontsize=12)
plt.legend(title='Sport', bbox_to_anchor=(1, 1), loc='upper left', fontsize='small')
plt.grid(True)
plt.show()
```

```python
# Pivot data untuk membuat stacked bar chart
pivot_table = top_3_sports_by_year.pivot(index='Year', columns='Sport', values='Earnings ($ million)').fillna(0)

# Membuat stacked bar chart
pivot_table.plot(kind='bar', stacked=True, figsize=(10, 6))

# Menambahkan label pada sumbu x dan y
plt.xlabel("Year", fontsize=12)
plt.ylabel("Total Earnings ($ million)", fontsize=12)

# Menambahkan judul
plt.title("Top 3 Sports with Highest Earnings by Year", fontsize=15)

# Menampilkan legenda
plt.legend(title="Sport", title_fontsize=12)

# Menampilkan grafik
plt.show()
```

```python
# Grouping berdasarkan kolom 'Name' dan 'Sport' dan menjumlahkan pendapatan mereka
total_earnings = df.groupby('Name')['Earnings ($ million)'].sum()

# Mengurutkan data berdasarkan total pendapatan secara menurun (descending)
sorted_earnings = total_earnings.sort_values(ascending=False)

# Mengambil 10 atlet dengan penghasilan terbanyak
top_10_athletes = sorted_earnings.head(10)

# Menampilkan hasil
top_10_athletes
```

```python
# Convert the Series to a DataFrame
top_10_df = top_10_athletes.reset_index()

# Rename the column to remove parentheses and spaces
top_10_df.rename(columns={'Earnings ($ million)': 'Earnings'}, inplace=True)

# Plot the bar plot
plt.figure(figsize=(10, 5))
sns.barplot(data=top_10_df, y="Name", x='Earnings', orient='h')
plt.title("Top 10 Highest Earnings (Forbes Richest Athletes 1990-2020)", fontsize=20)
plt.xlabel('Earnings ($ million)', fontsize=15)
plt.ylabel("Name", fontsize=15)
plt.show()
```

```python
top_athletes_by_year = df.loc[df.groupby('Year')['Earnings ($ million)'].idxmax()]
top_athletes_by_year
```

```python
# Ambil atlit dengan earnings tertinggi untuk setiap tahun
top_athletes_by_year = df.groupby('Year').apply(lambda x: x.nlargest(1, 'Earnings ($ million)'))

# Hitung total earnings seluruh atlit untuk setiap tahun
total_earnings_by_year = df.groupby('Year')['Earnings ($ million)'].sum().reset_index()

# Reset the index for top_athletes_by_year DataFrame
top_athletes_by_year.reset_index(drop=True, inplace=True)

# Gabungkan data atlit tertinggi dan total earnings
combined_data = pd.merge(total_earnings_by_year, top_athletes_by_year[['Year', 'Earnings ($ million)']], on='Year', suffixes=('_Total', '_Top'))

# Buat stacked barchart
plt.figure(figsize=(10, 6))
plt.bar(combined_data['Year'], combined_data['Earnings ($ million)_Total'], label='Total Earnings', color='b')
plt.bar(combined_data['Year'], combined_data['Earnings ($ million)_Top'], label='Top Athlete Earnings', color='r', bottom=combined_data['Earnings ($ million)_Total'])
plt.xlabel('Year', fontsize=12)
plt.ylabel('Earnings ($ million)', fontsize=12)
plt.title('Total Earnings vs Top Athlete Earnings per Year', fontsize=15)
plt.legend()
plt.show()
```

```python
athletes = df.groupby('Name')
df_appearances = athletes['Name'].count()\
                .reset_index(name='Number entered into the list')\
                .sort_values(['Number entered into the list'], ascending=0)
df_appearances = df_appearances[df_appearances['Number entered into the list'] >= 0]
df_appearances_top_10 = df_appearances.head(10)
df_appearances_top_10
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df_appearances_top_10, x='Number entered into the list', y='Name')
plt.title("Top 10 Athletes with Most Appearances in Forbes Richest Athletes (Forbes Richest Athletes 1990-2020)", fontsize=15)
plt.xlabel('Number of Appearances', fontsize=15)
plt.xticks(rotation=45)
plt.ylabel("Name", fontsize=15)
plt.yticks(fontsize=12)
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(integer=True))

plt.show()
```

```python
# Convert the Series objects to DataFrames
total_earnings = total_earnings.reset_index()
df_appearances = df_appearances.reset_index()

athlete_info = pd.merge(total_earnings, df_appearances, on='Name')

# Sort the athletes based on 'Earnings ($ million)' (descending) and 'Number entered into the list' (ascending)
athlete_info['Average Earnings'] = athlete_info['Earnings ($ million)'] / athlete_info['Number entered into the list']
sorted_athletes = athlete_info.sort_values(by='Average Earnings', ascending=False)

# Print the athlete with the highest earnings and the least number of appearances
top_athlete = sorted_athletes.iloc[0]
print("Athlete with the highest earnings and least number of appearances:")
print(top_athlete)
```
